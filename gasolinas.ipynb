```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


gas = pd.read_excel('transaccionesGas.xlsx')
gas['Fecha'] = pd.to_datetime(gas['Fecha'],format='%d/%m/%Y %H:%M:%S')
```

```python
gas = gas[gas['Tipo / Naturaleza'] == 'Venta / Estándar']
gas = gas[(gas['Cantidad']<60) & (gas['Importe']>20) & (gas['Cantidad']>0)]
```

# Análisis exploratorio

### Transacciones por fecha

```python
transacciones_por_fecha = gas['Fecha'].dt.date.value_counts().sort_index()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(transacciones_por_fecha.index, transacciones_por_fecha.values, width=0.8, align='center')
plt.title('Transacciones por Fecha')
plt.xlabel('Fecha')
plt.ylabel('Número de Transacciones')
plt.grid(axis='y')

# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.plot(transacciones_por_fecha.index, transacciones_por_fecha.values, marker='o', linestyle='-', color='b')
plt.title('Transacciones por Fecha')
plt.xlabel('Fecha')
plt.ylabel('Número de Transacciones')
plt.grid(axis='y')

# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

### Transacciones por dia de la semana

```python
transacciones_por_dia_semana = gas['Fecha'].dt.dayofweek
# Crear un diccionario para mapear números de día de la semana a nombres
dias_semana = {
    0: 'Lunes',
    1: 'Martes',
    2: 'Miércoles',
    3: 'Jueves',
    4: 'Viernes',
    5: 'Sábado',
    6: 'Domingo'
}
```

```python
# Calcular el promedio de transacciones por día de la semana
promedio_transacciones = transacciones_por_dia_semana.value_counts() / transacciones_por_dia_semana.nunique()
# Encontrar el día de la semana con el promedio más alto
dia_con_mas_transacciones_promedio = dias_semana[promedio_transacciones.idxmax()]
# Imprimir el resultado
print(f"En promedio, el día de la semana con más transacciones es {dia_con_mas_transacciones_promedio}.")
```

```python
# Calcular el promedio de transacciones por día de la semana
promedio_transacciones = transacciones_por_dia_semana.value_counts() / transacciones_por_dia_semana.nunique()
# Encontrar el día de la semana con el promedio más alto
dia_con_mas_transacciones_promedio = dias_semana[promedio_transacciones.idxmin()]
# Imprimir el resultado
print(f"En promedio, el día de la semana con menos transacciones es {dia_con_mas_transacciones_promedio}.")
```

```python

# Agrupar transacciones por día de la semana (0 = Lunes, 1 = Martes, ..., 6 = Domingo)
consumo_por_dia_semana = gas.groupby(gas['Fecha'].dt.dayofweek).size()

# Calcular el promedio de transacciones por día de la semana
promedio_transacciones = consumo_por_dia_semana / consumo_por_dia_semana.groupby(gas['Fecha'].dt.month).transform('count')

# Renombrar los días de la semana
dias_semana = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
promedio_transacciones.index = [dias_semana[dia] for dia in promedio_transacciones.index]

# Crear el gráfico de líneas por hora
plt.figure(figsize=(10, 6))
plt.plot(promedio_transacciones.index, promedio_transacciones.values, marker='o', linestyle='-', color='b')
plt.title('Número de Transacciones por Día de la semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Promedio de Transacciones')
plt.grid(True)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

### Tiempo promedio de espera

#### Tiempo promedio 

```python
# Ordenar los datos por fecha
gas.sort_values(by='Fecha', inplace=True)

# Calcular el tiempo entre transacciones
gas['Tiempo_Entre_Transacciones'] = gas['Fecha'].diff()

# Calcular el tiempo promedio entre transacciones
tiempo_promedio = gas['Tiempo_Entre_Transacciones'].mean()

# Calcular el tiempo máximo entre transacciones
tiempo_maximo = gas['Tiempo_Entre_Transacciones'].max()

# Visualizar el tiempo entre transacciones
plt.figure(figsize=(12, 6))
plt.plot(gas['Fecha'], gas['Tiempo_Entre_Transacciones'], marker='o', linestyle='-')
plt.title('Tiempo Entre Transacciones en la Gasolinera')
plt.xlabel('Fecha y Hora')
plt.ylabel('Tiempo Entre Transacciones')
plt.grid(True)
plt.show()

print(f"Tiempo promedio entre transacciones: {tiempo_promedio}")
print(f"Tiempo máximo entre transacciones: {tiempo_maximo}")
```

#### Tiempo promedio por isla

```python
gas = gas.sort_values(by=['TPV', 'Fecha'])
# Calcular el tiempo de espera para cada transacción
gas['Tiempo_Entre_Transacciones'] = gas.groupby('TPV')['Fecha'].diff()
# Usar pivot_table para calcular el tiempo de espera promedio por isla de servicio
tiempo_espera_por_isla = gas.pivot_table(index='Fecha', columns='TPV', values='Tiempo_Entre_Transacciones', aggfunc='mean')
```

```python
tiempo_espera_por_isla.describe()
```

```python
transacciones_menor_5min = gas[gas['Tiempo_Entre_Transacciones'] < pd.Timedelta(minutes=5)]
tiempo_espera_menor_5min = transacciones_menor_5min.pivot_table(index='Fecha', columns='TPV', values='Tiempo_Entre_Transacciones', aggfunc='mean')
tiempo_espera_menor_5min.describe()
```

### Consumidor promedio

```python
gas['Date'] = gas['Fecha'].dt.date
resumen_diario = gas.groupby(['Date','Artículo' ])['Cantidad'].sum().unstack().reset_index()

# Crear gráfico de líneas
plt.figure(figsize=(12, 6))
for producto in ['MAGNA', 'PREMIUM', 'DIESEL']:
    plt.plot(resumen_diario['Date'], resumen_diario[producto], label=producto)

# Personalizar el gráfico
plt.title('Consumo de Combustible por Producto (Resumen Diario)')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Combustible')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
